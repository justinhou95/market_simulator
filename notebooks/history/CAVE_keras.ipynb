## Simple implementation of CVAE with keras

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import utils
from sklearn.datasets import make_circles
from sklearn.preprocessing import MinMaxScaler
from tqdm.auto import tqdm
import tensorflow as tf
from importlib import reload
from tensorflow import keras
from tensorflow.keras import layers
import base
from utils import tosig
import cvae_keras
reload(cvae_keras)
```

```python
data, conditions = make_circles(n_samples=10000, noise=0.05, factor=0.5)
scaler = MinMaxScaler()
data = scaler.fit_transform(data)
# Plot dataset
plt.scatter(*data.T, c=conditions, s=2)
plt.show()
data = utils.as_float_array(data)
data_cond = utils.as_float_array(conditions)
data_cond = data_cond.reshape(-1, 1)
```

## CVAE

```python
model = cvae_keras.CVAE(n_latent = 4, n_hidden=50, alpha = 0.02)
model.build(data.shape[-1],data_cond.shape[-1])
model.train(data,data_cond,n_epochs=5000)
```

```python
n_samples = 1000

randoms = np.random.normal(0, 1, size=(n_samples, 4))
condi0 = np.zeros(shape=(n_samples, 1))
outputs0 = model.decoder([tf.constant(randoms), tf.constant(condi0)]).numpy()

randoms = np.random.normal(0, 1, size=(n_samples, 4))
condi1 = np.zeros(shape=(n_samples, 1)) + 1
outputs1 = model.decoder([tf.constant(randoms), tf.constant(condi1)]).numpy()

plt.scatter(*outputs0.T, s=2)
plt.scatter(*outputs1.T, s=2)

plt.show()
```

## VAE

```python
model2 = cvae_keras.VAE(n_latent = 4, n_hidden=50, alpha = 0.02)
model2.build(data.shape[-1])
model2.train(data,n_epochs=5000)
```

```python
n_samples = 1000
randoms = np.random.normal(0, 1, size=(n_samples, 4))
outputs = model2.decoder(tf.constant(randoms)).numpy()
plt.scatter(*outputs.T, s=2, alpha = 1)
plt.scatter(*data.T, s=2, alpha = 0.5)
```
