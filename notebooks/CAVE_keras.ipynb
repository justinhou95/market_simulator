```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import utils
from sklearn.datasets import make_circles
from sklearn.preprocessing import MinMaxScaler
from tqdm.auto import tqdm
import tensorflow as tf
import importlib
from tensorflow import keras
from tensorflow.keras import layers
```

## Generate Data

```python
data, conditions = make_circles(n_samples=10000, noise=0.05, factor=0.5)
# We normalise the data to [0, 1]^2
scaler = MinMaxScaler()
data = scaler.fit_transform(data)
# Plot dataset
plt.scatter(*data.T, c=conditions, s=2)
plt.show()
```

## Preprocessing Data

```python
data = utils.as_float_array(data)
data_cond = utils.as_float_array(conditions)
data_cond = data_cond.reshape(-1, 1)
```

## Build Model

```python
tf.keras.backend.set_floatx('float64')
n_latent=4
n_hidden=50
input_dim = data.shape[1]
cond_dim = data_cond.shape[1]

# Sampling
class Sampling(layers.Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = keras.backend.random_normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

# Encoder 
inputs_data = keras.Input(shape=(input_dim,))
inputs_cond = keras.Input(shape=(cond_dim,))

x = layers.concatenate([inputs_data,inputs_data],axis = 1)

x = layers.Dense(units=n_hidden)(x)
x = layers.LeakyReLU(alpha=0.3)(x)

mn = layers.Dense(units=n_latent)(x)
mn = layers.LeakyReLU(alpha=0.3)(mn)

sd = layers.Dense(units=n_latent)(x)
sd = layers.LeakyReLU(alpha=0.3)(sd)

z = Sampling()([mn, sd])

encoder = keras.models.Model([inputs_data, inputs_cond], [mn, sd, z])
# encoder.summary()

# Decoder 
inputs_z = keras.Input(shape=(n_latent,))
inputs_label = keras.Input(shape=(cond_dim,))
x = layers.concatenate([inputs_z, inputs_label], axis=1)

x = layers.Dense(units = n_hidden)(x)
x = layers.LeakyReLU(alpha=0.3)(x)
x = layers.Dense(units = input_dim, activation = 'sigmoid')(x)

decoder = keras.models.Model([inputs_z, inputs_label], x)
# decoder.summary()
```

```python
# CVAE
inputs_data = keras.Input(shape=(input_dim,))
inputs_cond = keras.Input(shape=(cond_dim,))
mn, sd, z = encoder([inputs_data, inputs_cond])
x = decoder([z, inputs_cond])
alpha = 0.02
reconstruction = x
decoded_loss = tf.reduce_sum(tf.square(reconstruction - inputs_data), axis = -1, keepdims = True)
latent_loss = -0.5 * tf.reduce_sum(1. + sd - tf.square(mn) - tf.exp(sd), axis = -1, keepdims = True)
loss = (1 - alpha) * decoded_loss + alpha * latent_loss
cvae = keras.models.Model([inputs_data, inputs_cond], loss)
```

## Train Model

```python
cvae.compile(optimizer='adam', loss='mae')
cvae.fit(x = [data, data_cond], y = np.zeros(shape= (10000,1)), epochs=5000, batch_size=data.shape[0], verbose = False)
cvae.evaluate(x = [data, data_cond], y = np.zeros(shape= (10000,1)))
```
