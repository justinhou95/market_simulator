```python
# Actually for rough_bergomi model, conditional VAE does no help because the cond are independent

# rough_bergomi
params = {
            "M": 100,
            "H": 0.14,
            "rho": -0.85,
            "xi0": 0.1,
            "nu": 1.5,
            "S0": 1.
          }

def _load_rough_bergomi(params, test_use = False):
    grid_points_dict = {"M": 28, "W": 5, "Y": 252}
    grid_points = grid_points_dict['M']
    params["T"] = grid_points / grid_points_dict["Y"]
    paths = rough_bergomi(grid_points, **params)
    if test_use:
        windows = paths
    else:   
        windows = [leadlag(path) for path in paths]
    return windows

windows = _load_rough_bergomi(params)

for path in windows:
    returns = path[::2, 1] / path[0, 1]
    plt.plot(returns, "b", alpha=0.1)

plt.title("Monthly paths")
plt.xlabel("Days")
plt.show()
```

## Log-signature computing

```python
order = 4
def _logsig(path):
    return esig.tosig.stream2logsig(path, order)
def _sig(path):
    return esig.tosig.stream2sig(path, order)
orig_logsig = np.array([_logsig(path) for path in tqdm(windows, desc="Computing log-signatures")])
orig_sig = np.array([_sig(path) for path in tqdm(windows, desc="Computing signatures")])
scaler = MinMaxScaler(feature_range=(0.00001, 0.99999))
logsig = scaler.fit_transform(orig_logsig)
logsigs = logsig[1:]   # 1 week forecasting 1 week 
conditions = logsig[:-1] 
```

## From Signature to Path

1. Glue the condition and path together 

```python
import logsig_inversion
from esig.tosig import stream2logsig
from utils.leadlag import leadlag
from gpsig import kernels
import gpflow
```

```python
def TU2(Kxx,Kxy,Kyy):
    m = Kxx.shape[0]
    n = Kyy.shape[0]
    TU = 0.
    TU += (np.sum(Kxx) - np.trace(Kxx)) / (m * (m-1))
    TU += (np.sum(Kyy) - np.trace(Kyy)) / (n * (n-1))
    TU -= 2 * np.sum(Kxy) / (m * n)
    return TU

def sig_TU2(X,Y):
    num_levels = 4
    num_lags = 4
#     kern = kernels.SignatureKernel(base_kernel = gpflow.kernels.SquaredExponential(), num_levels = num_levels, order = num_levels, num_lags = num_lags)
    kern = kernels.SignatureKernel(base_kernel = gpflow.kernels.Exponential(), num_levels = num_levels, order = num_levels, num_lags = num_lags)
    Kxx = kern.K(X,X)
    Kxy = kern.K(X,Y)
    Kyy = kern.K(Y,Y)
    return TU2(Kxx,Kxy,Kyy)
```

```python
pip = 0.001
n_pips = 5 * 10
n_points = 28
n_iterations = 100
n_organisms = 100
recovered_paths = []
for i, logsig in enumerate(generated):
    recovered_paths.append(logsig_inversion.train(logsig, order, n_iterations, n_organisms, n_points, pip, n_pips)[0] + 1.)
    
```

```python
H = []
X = tf.constant(recovered_paths)[:,:,None]
params = {
            "M": 99,
            "H": 0.14,
            "rho": -0.85,
            "xi0": 0.1,
            "nu": 1.5,
            "S0": 1.
          }
for i in range(10):
    Y = tf.constant(_load_rough_bergomi(params, test_use = True))[:,:,None]
    H.append(sig_TU2(X,Y))
```

```python
H0 = []
params = {
            "M": 99,
            "H": 0.14,
            "rho": -0.85,
            "xi0": 0.1,
            "nu": 1.5,
            "S0": 1.
          }
for i in range(10):
    X = tf.constant(_load_rough_bergomi(params, test_use = True))[:,:,None]
    Y = tf.constant(_load_rough_bergomi(params, test_use = True))[:,:,None]
    H0.append(sig_TU2(X,Y))
```

```python
H1 = []
params_prime = {
            "M": 99,
            "H": 0.14,
            "rho": -0.85,
            "xi0": 1,
            "nu": 1.5,
            "S0": 1.
          }
for i in range(10):
    X = tf.constant(_load_rough_bergomi(params, test_use = True))[:,:,None]
    Y = tf.constant(_load_rough_bergomi(params_prime, test_use = True))[:,:,None]
    H1.append(sig_TU2(X,Y))
```
