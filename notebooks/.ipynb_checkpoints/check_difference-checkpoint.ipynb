```python
import base
import numpy as np
from tqdm.auto import tqdm
from utils import tosig
from utils.leadlag import leadlag
from rough_bergomi import rough_bergomi
from importlib import reload
import matplotlib.pyplot as plt

import evaluation
reload(evaluation)

import process_discriminator
reload(process_discriminator)


import gpflow

import tensorflow as tf
from gpsig import kernels
%matplotlib inline
import esig
tf.keras.backend.set_floatx('float64')
reload(kernels)
```

```python
def _load_rough_bergomi(params):
    grid_points_dict = {"M": 28, "W": 5, "Y": 252}
    grid_points = grid_points_dict['M']
    params["T"] = grid_points / grid_points_dict["Y"]
    paths = rough_bergomi(grid_points, **params)
    windows = [leadlag(path) for path in paths]
    return windows

def _load_rough_bergomi_pure(params):
    grid_points_dict = {"M": 28, "W": 5, "Y": 252}
    grid_points = grid_points_dict['M']
    params["T"] = grid_points / grid_points_dict["Y"]
    paths = rough_bergomi(grid_points, **params)
    windows = paths
    return windows
```

```python
order = 3
params1 = {
            "M": 100,
            "H": 0.14,
            "rho": -0.85,
            "xi0": 0.1,
            "nu": 1.5,
            "S0": 1.
          }
params2 = {
            "M": 100,
            "H": 0.14,
            "rho": -0.85,
            "xi0": 0.5,
            "nu": 1.5,
            "S0": 1.
          }

windows1 = np.array(_load_rough_bergomi_pure(params1))
windows2 = np.array(_load_rough_bergomi_pure(params2))
plt.plot(windows1.T, "g", alpha=0.05)
plt.plot(windows2.T, "r", alpha=0.05)
plt.show()
```

## Comparison of different kernels with small data set (m = 100)

```python
num_levels = 3
num_lags = 0
kernE = kernels.SignatureKernel(base_kernel = gpflow.kernels.Exponential(), \
                                num_levels = num_levels, order = num_levels, num_lags = num_lags)
```

```python
TU1 = []
TU0 = []
TU1_soft = []
TU0_soft = []
TU1_hard = []
TU0_hard = []
TU1_soft_fix = []
TU0_soft_fix = []
TU1_hard_fix = []
TU0_hard_fix = []

X = tf.constant(_load_rough_bergomi(params1))
Y = tf.constant(_load_rough_bergomi(params2))
phi = process_discriminator.T_global(X, Y, order=3, verbose=True, normalise=True, compute_sigs=True)
phi_hard = evaluation.T_global(X, Y, order=3, verbose=True, normalise=True, compute_sigs=True)

for i in range(20):
    X = tf.constant(_load_rough_bergomi(params1))
    Y = tf.constant(_load_rough_bergomi(params2))
    Z = tf.constant(_load_rough_bergomi(params1))
    
    result, TU = process_discriminator.test(X, Y, order=3, confidence_level=0.99, normalise = False)
    TU1.append(TU)
    result, TU = process_discriminator.test(X, Z, order=3, confidence_level=0.99, normalise = False)
    TU0.append(TU)
    
    result, TU = process_discriminator.test(X, Y, order=3, confidence_level=0.99)
    TU1_soft.append(TU)
    result, TU = process_discriminator.test(X, Z, order=3, confidence_level=0.99)
    TU0_soft.append(TU)
    
    result, TU = evaluation.test(X, Y, order=3, confidence_level=0.99)
    TU1_hard.append(TU)
    result, TU = evaluation.test(X, Z, order=3, confidence_level=0.99)
    TU0_hard.append(TU)
    
    result, TU = process_discriminator.test_fix(X, Y, order=3, confidence_level=0.99, phi_x = phi)
    TU1_soft_fix.append(TU)
    result, TU = process_discriminator.test_fix(X, Z, order=3, confidence_level=0.99, phi_x = phi)
    TU0_soft_fix.append(TU)
    
    result, TU = evaluation.test_fix(X, Y, order=3, confidence_level=0.99, phi_x = phi_hard)
    TU1_hard_fix.append(TU)
    result, TU = evaluation.test_fix(X, Z, order=3, confidence_level=0.99, phi_x = phi_hard)
    TU0_hard_fix.append(TU)
    
    
```

```python
H1_E = []
H0_E = []

for i in range(20):
    print(i)
    X = tf.constant(_load_rough_bergomi(params1))
    Y = tf.constant(_load_rough_bergomi(params2))
    Z = tf.constant(_load_rough_bergomi(params1))
    H1_E.append(evaluation.Sig_TU2(X,Y,kernE))
    H0_E.append(evaluation.Sig_TU2(X,Z,kernE))
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(2,3,1)
plt.hist(TU0,alpha = 0.5)
plt.hist(TU1,alpha = 0.5)
plt.title('no normalization')
plt.legend(['Same','Different'])
plt.subplot(2,3,2)
plt.hist(TU0_soft,alpha = 0.5)
plt.hist(TU1_soft,alpha = 0.5)
plt.title('soft')
plt.legend(['Same','Different'])
plt.subplot(2,3,3)
plt.hist(TU0_soft_fix,alpha = 0.5)
plt.hist(TU1_soft_fix,alpha = 0.5)
plt.title('soft_fix')
plt.legend(['Same','Different'])
plt.subplot(2,3,4)
plt.hist(TU0_hard,alpha = 0.5)
plt.hist(TU1_hard,alpha = 0.5)
plt.title('hard')
plt.legend(['Same','Different'])
plt.subplot(2,3,5)
plt.hist(TU0_hard_fix,alpha = 0.5)
plt.hist(TU1_hard_fix,alpha = 0.5)
plt.title('hard_fix')
plt.legend(['Same','Different'])
plt.subplot(2,3,6)
plt.hist(H0_E,alpha = 0.5)
plt.hist(H1_E,alpha = 0.5)
plt.title('Exponential')
plt.legend(['Same','Different'])
plt.show()
```
