```python
def _load_rough_bergomi(params):
    grid_points_dict = {"M": 28, "W": 5, "Y": 252}
    grid_points = grid_points_dict['M']
    params["T"] = grid_points / grid_points_dict["Y"]
    paths = rough_bergomi(grid_points, **params)
    windows = [leadlag(path) for path in paths]
    return windows

def _load_rough_bergomi_pure(params):
    grid_points_dict = {"M": 28, "W": 5, "Y": 252}
    grid_points = grid_points_dict['M']
    params["T"] = grid_points / grid_points_dict["Y"]
    paths = rough_bergomi(grid_points, **params)
    windows = paths
    return windows
```

```python
order = 3
params1 = {
            "M": 100,
            "H": 0.14,
            "rho": -0.85,
            "xi0": 0.1,
            "nu": 1.5,
            "S0": 10.
          }
params2 = {
            "M": 100,
            "H": 0.14,
            "rho": -0.85,
            "xi0": 0.2,
            "nu": 1,
            "S0": 10.
          }

windows1 = np.array(_load_rough_bergomi_pure(params1))
windows2 = np.array(_load_rough_bergomi_pure(params2))



plt.figure(figsize = [8,4])
plt.plot(windows1.T, color = 'tab:blue', alpha=0.1)
plt.plot(windows2.T, color = 'tab:orange', alpha=0.1)
plt.legend(['$V_{0} = 0.3$', '$V_{0} = 0.5$'])
plt.grid()
plt.show()
```
